## 1.2  그래프 그리기
1. plt.figure(figsize=(10, 6)) : 도화지 펼친다   
2. plt.plot() : 그래프를 그린다   
3. plt.show : 시각화한다   

```python
# 도화지를 펼친다
plt.figure(figsize=(10, 6))

# 그래프를 그린다
plt.plot([0, 1, 2, 3, 4, 5], [1, 7, 2, 5, 4, 4])

# 시각화한다
plt.show()
```

- 삼각함수 그리기

```python
x = np.linspace(0, 10, 100)

# 도화지를 펼친다
plt.figure(figsize=(10, 6))

# 도화지를 그린다
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.tan(x))

# 시각화 한다
plt.show()
```

**option1. color, linestyle, marker**

* plt.plot(x, y, color='red') -> red, green, blue 등 색상 이름
* plt.plot(x, y, color='r') -> r, g, b, y, m(자홍), k(검정) 등
* plt.plot(x, y, color='0.2') -> 회색조(0-1사이 값)
* plt.plot(x, y, color='#ff0000') -> 16진수 색상 값
* plt.plot(x, y, linestyle='solid') -> 실선('-')
* plt.plot(x, y, linestyle='dashed') -> 파선('--')
* plt.plot(x, y, linestyle='dashdot') -> 1점 쇄선('-.')
* plt.plot(x, y, linestyle='dotted') -> 점선(':')
* plt.plot(x, y, '--r') -> 빨간색 파선   

```python
# 도화지를 펼친다
plt.figure(figsize=(10, 6))

# 도화지를 그린다
plt.plot(x, np.sin(x), '-r')
plt.plot(x, np.cos(x), '--b')
plt.plot(x, np.tan(x), '--^g')

# 시각화 한다
plt.show()
```

```python
# 도화지를 펼친다
plt.figure(figsize=(10, 6))

# 그래프를 그린다
plt.plot(x, np.sin(x), '-r')
plt.plot(x, np.cos(x), '--b')
plt.plot(x, np.tan(x), '--^g')

# 시각화 한다
plt.show()
```

```python
y = np.sin(x)

# 도화지를 펼친다
plt.figure(figsize=(10, 6))

# 그래프를 그린다
plt.plot(
    x,
    y,
    color="red", 
    linestyle="--",
    marker="o", # o,<,> 
    markerfacecolor="blue",
    markersize=5, 
)

# 시각화 한다.
plt.show()
```

```python
# 도화지를 펼친다
plt.figure(figsize=(10, 6))

# 그래프를 그린다
for m in [ '+', ',', '.', '1', '2', '3', '4', 'o', 'x', '+', 'v', '^', '<', '>', 's', 'd', 'p']:
    plt.plot(np.random.randint(20), 
             np.random.randint(20), 
             m,
             markersize=10,
             label=f'marker={m}')

# 시각화 한다
plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1.05), fancybox=True, shadow=True)
plt.show()
```

```python
# 도화지를 펼친다
plt.figure(figsize=(10, 6))

# 그래프를 그린다
plt.plot(x, np.sin(x), '-r', label='sin') # red --
plt.plot(x, np.cos(x), '--b', label='cos') # blue square
plt.plot(x, np.tan(x), '--^g', label='tan')

# 옵션

plt.legend(loc='upper right', bbox_to_anchor=(1, 1.05),
           ncol=2, fancybox=True, shadow=True)
# bbox_to_anchor : 박스 위치
# ncol : 컬럼의 수
# fancybox : 모서리 둥글게
# shadow : 그림자

plt.grid(True) # 격자무늬 
plt.title("Sample") # 제목
plt.xlabel("x_title") # x축 제목
plt.ylabel("y_title") # y축 제목 

# 시각화 한다
plt.show()
```

**option4. 경계 표현**
* 자동으로 표현
* plt.xlim, plt.ylim으로 상세 조정 가능
* plt.axis([x축 최솟값, x축 최댓값, y축 최솟값, y축 최댓값])로 한 번에 조정 가능
* plt.axis('keyword') -> tight, equal(균등)

```python
# 도화지를 펼친다
plt.figure(figsize=(10, 6))

# 그래프를 그린다
plt.plot(x, np.sin(x), '-r', label='sin') # red --
plt.plot(x, np.cos(x), '--b', label='cos') # blue square
plt.plot(x, np.tan(x), '--^g', label='tan')

# 옵션
plt.grid(True) # 격자무늬 
plt.legend(loc='upper right') # 범례(위치)
plt.title("Sample") # 제목
plt.xlabel("x_title") # x축 제목
plt.ylabel("y_title") # y축 제목 

# 경계표현
# plt.xlim(1, 10)
# plt.ylim(-10, 10)
plt.axis([1, 10, -10, 10])
# plt.axis('tight')
# plt.axis('equal')

# 시각화 한다
plt.show()
```

**option5. 기타 attribute**

* alpha : 투명도 
* logy : Y축 Log scaling  
* kind : line, bar, barh, kde
* subplots : 여러개의 plot 그리기
* legend : subplot의 범례 지정
* xlim, ylim : X축과 Y축의 경계(axis로 한번에 그릴 수 있음)
* grid : 그리드 표현(True, False)
* title : 그래프의 제목 지정 
* linewidth : 라인 넓이
* color : 색
* linestyle : 실선, 점선, 1점 쇄선 등
* marker : 마커 지정
* markerfacecolor : 마커 색
* markersize : 마커 크기
* 그 외 Attribute : https://matplotlib.org/3.1.1/tutorials/introductory/pyplot.html

## 1.5 막대그래프(bar chart)
- 범주형 데이터의 분포를 파악한다.   
- 데이터프레임에서 바로 그릴 수 있다.   
- 일반적으로 데이터가많은 경우 정렬 후 그리는 것이 효과적이다.   

```python
plt.bar(['A', 'B', 'C'], [4, 2, 3])

plt.title('bar chart', fontsize=15)
plt.xlabel('x_title', fontsize=15)
plt.ylabel('y_title', fontsize=15)

plt.show()
```

```python
plt.barh(['A', 'B', 'C'], [4, 2, 3])

plt.title('bar chart', fontsize=15)
plt.xlabel('x_title', fontsize=15)
plt.ylabel('y_title', fontsize=15)

plt.show()
```

```python
plt.bar(['A', 'B', 'C'], [4, 2, 3], color='g', label='BarChart1') # 라벨링 하는 방법 1
plt.bar(['A', 'B', 'C'], 
        [2, 5, 3], 
        color='b', 
        alpha=0.5, 
        bottom=[4, 2, 3],
        label= 'BarChart2')

plt.title('bar chart', fontsize=15)
plt.xlabel('x_title', fontsize=15)
plt.ylabel('y_title', fontsize=15)

plt.legend()

plt.show()
```

```python
plt.bar(['A', 'B', 'C'], [4, 2, 3], color='g')
plt.bar(['A', 'B', 'C'], 
        [2, 5, 3], 
        color='b', 
        alpha=0.5, 
        bottom=[4, 2, 3],
)

plt.title('bar chart', fontsize=15)
plt.xlabel('x_title', fontsize=15)
plt.ylabel('y_title', fontsize=15)

plt.legend(['BarChart1', 'BarChart2']) # 라벨링 하는 방법 2

plt.show()
```

## 1.7 subplot

```python
x = np.linspace(0, 10, 100)
y = x ** 2
```

```python
# 1개의 row, 2개의 columns, 그 중 n번째

plt.subplot(121)
plt.plot(x, y, 'r')

plt.subplot(122)
plt.plot(x, y, 'g')

plt.show()
```

```python
plt.subplot(231)
plt.plot(x, y, 'r')

plt.subplot(232)
plt.plot(x, y, 'g')

plt.subplot(233)
plt.plot(x, y, 'b')

plt.subplot(234)
plt.plot(x, y, 'r')

plt.subplot(235)
plt.plot(x, y, 'g')

plt.subplot(236)
plt.plot(x, y, 'b')

plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r')

plt.subplot(1,2,2)
plt.plot(x, y, 'g')

plt.show()
```

```python
grid = plt.GridSpec(2, 3)

plt.subplot(grid[0, 0])
plt.plot(x, y, 'r')

plt.subplot(grid[0, 1:])
plt.plot(x, y, 'g')

plt.subplot(grid[1, :2])
plt.plot(x, y, 'r')

plt.subplot(grid[1, 2])
plt.plot(x, y, 'g')

plt.show()
```
