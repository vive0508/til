### 2.2  그래프 그리기

- 일반적인 그래프 그리기

```python
x = np.linspace(0, 10, 100)
y1 = x**2
y2 = x**2.5
y3 = x**3
```

```python
# 도화지를 깐다
plt.figure(figsize=(5, 5))

# 그림을 그린다
plt.plot(x, y1, x, y2, x, y3)

# 시각화
plt.show()
```

- sns.set_style()

```python
# sns.set_style()
# "white", "whitegrid", "dark", "darkgrid", "ticks"

sns.set_style("darkgrid")
plt.figure(figsize=(5, 5))
plt.plot(x, y1, x, y2, x, y3) 
plt.show()
```

- sns.despine()

```python
# despine()으로 왼쪽, 아랫쪽 선만 남겨놓을 수 있다
# offset으로 왼쪽, 아랫쪽 선의 간격을 조정할 수 있다(옵션)

sns.set_style("white")
plt.figure(figsize=(5, 5))
plt.plot(x, y1, x, y2, x, y3) 
sns.despine(offset=10)
plt.show()
```

### 2.4 다양한 형태의 그래프 그리기

### 2.4.1 boxplot

- [seaborn boxplot 공식문서](https://seaborn.pydata.org/generated/seaborn.boxplot.html?highlight=boxplot#seaborn.boxplot)
- Median 중심으로 아래 위 상자가 각각 전체 데이터 25%씩 총 50%를 가진다.   
- 상자 전체 길이를 IQR(Inter Quatile Range)라고 한다.   
- IQR의 1.5배 이상을 벗어나면 outlier로 점으로 별도 표기한다.   

- box plot 기본

```python
plt.figure(figsize=(5, 5))
sns.boxplot(x=tips["total_bill"])
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

- option

```python
# hue : 카테고리 데이터 표현
# palette : 색상 표현 (Set1~Set3)
plt.figure(figsize=(5, 5))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3", linewidth=1)
plt.show()
```

- swarmplot과 함께 그리기

```python
# color: 0~1 사이 검은색부터 흰색 사이 값을 조절 
plt.figure(figsize=(5, 5))
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
plt.show()
```

### 2.4.3 lmplot
> [seaborn lmplot 공식문서](https://seaborn.pydata.org/generated/seaborn.lmplot.html?highlight=seaborn%20lmplot#seaborn.lmplot)   
> 색칠된 데이터의 간격이 좁을 수록 강한 상관관계를 갖는다.

- lmplot 기본

```python
plt.figure(figsize=(5, 5))
sns.lmplot(x="total_bill", y="tip", data=tips)
plt.show()
```

- option

```python
# query라는 매서드로 I에 해당하는 데이터셋만 가져옴
# ci 신뢰구간 선택, None 옵션은 신뢰구간 영역 보이는 옵션을 끄는 것
# order : 차수에 따라 옵션 변경 
# robust : 원본에서 많이 떨어진 데이터(outlier)는 없는 셈 친다
# scater_kws 마커사이즈 변경

plt.figure(figsize=(5, 5))
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, markers=["o", "x"])
plt.show()
```

### 2.4.5 heatmap
> [seaborn heatmap 공식문서](https://seaborn.pydata.org/generated/seaborn.heatmap.html?highlight=seaborn%20heatmap#seaborn.heatmap)

- 데이터셋 불러오기

```python
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
flights.head()
```

```python
flights.info()
```

- heatmap 기본

```python
plt.figure(figsize=(5, 5))
sns.heatmap(flights)
plt.show()
```

- option

```python
# annot : 데이터 값 표시, fmt: 정수형 표현, linewidths : 간격, cmap : 색상
plt.figure(figsize=(5, 5))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, cmap="YlGnBu")
plt.show()
```
